```python
import csv 
import json
from urllib.request import urlopen, quote
import requests,csv
import pandas as pd #导入这些库后边都要用到

#================================     csv读取   ==============================

csv_file=csv.reader(open('pure_address.csv','r'))  #'r' read阅读模式
#print (csv_file)   #打印的只是一个对象的模型
#for add in csv_file:   #进行遍历打印
#    print (add)
    
#================================    csv 写入   ==============================

#Home=['My_Home',1000000]
#out = open('Add_short_testing.csv','a',newline='')


#csv_write=csv.writer(out,dialect='excel')

#csv_write.writerow(Home)

def getlatandlng(address):
    url = 'http://api.map.baidu.com/geocoder/v2/'
    output = 'json'
    ak = 'xNMFMZzuTF89G7wRysNkGIetwwAfA6FF'
    add = quote(address) #由于本文城市变量为中文，为防止乱码，先用quote进行编码
    uri = url + '?' + 'address=' + add  + '&output=' + output + '&ak=' + ak
    req = urlopen(uri)
    res = req.read().decode() #将其他编码的字符串解码成unicode
    temp = json.loads(res) #对json数据进行解析
    lat=temp["result"]["location"]["lat"]
    lng=temp["result"]["location"]["lng"]
    return [lat,lng]

for add in csv_file:   #进行遍历打印
    print (add)
    
    price=add[0]
    addressname=add[3]

    latandlng = getlatandlng(addressname)


    newlatandlng=[addressname,price,latandlng[0],latandlng[1]] 
    print (newlatandlng) 
    out = open('Add_short_testing_newnew.csv','a',newline='')
    csv_write=csv.writer(out,dialect='excel')
    csv_write.writerow(newlatandlng)
    
    
    
    
```
